```python
import numpy as np
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('./Files/ECV_2004_2018.csv.gz', sep = ';', compression = 'gzip')

print(df.shape)

df.head()
```

```python
alt.Chart(df3).mark_bar(color = 'green', size = 20).encode(
    alt.X('Year', title = 'A√±o',scale = alt.Scale(domain=(2003, 2018))),
    alt.Y('HHId', axis = alt.Axis(format='%', title = '(%) Poblaci√≥n'),scale = alt.Scale(domain=(0, 8))),
    tooltip=['Year', 'HHId']
).configure_axis(
    grid=False
).configure_view(
    strokeWidth=0
).properties(
    title = 'Evoluci√≥n de la poblaci√≥n riesgo de privaci√≥n material',
    width=600,
    height=250
)
```
